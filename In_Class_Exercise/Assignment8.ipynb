```
import Point
import copy
class Rectangle:
    def __init__(self, posn, w, h):
        self.corner = posn
        self.width = w
        self.height = h

    def __str__(self):
        return "({0},{1},{2})".format(self.corner, self.width, self.height)

    def grow(self,delta_width,delta_heigh):
        self.width+=delta_width
        self.height+=delta_heigh

    def move(self,dx,dy):
        self.corner.x+=dx
        self.corner.y+=dy

    def transpose(self):
        
        self.width, self.height = self.height, self.width

    def contains(self, point):
       
        x, y = point.get_x(), point.get_y()
        return 0 <= x < self.width and 0 <= y < self.height

box = Rectangle(Point.Point(0, 0), 100, 200)
bomb = Rectangle(Point.Point(100, 80), 5, 10)

box.grow(50,100)
box.move(100,100)

print("box:", box)
print("bomb:", bomb)
print(bomb.corner.x)

b1=Rectangle(Point.Point(1,2),100,200)
b2=copy.copy(b1)
b3=copy.deepcopy(b2)

print(b1.corner is b2.corner)
print(b1.corner is b3.corner)
```
